```python
import numpy as np
import pandas as pd

movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')                   
```

```python
# aynı başlığı olan tabloları birleştirdik
movies = movies.merge(credits,on="title")
```

```python
movies.head(1)
```

```python
# şimdi gereksiz sütunları kaldıralım
# film tavsiyesi verilmesi için gerekli sütunları alalım sadece:
movies=movies[["movie_id","title","overview","genres","keywords","cast","crew"]]
```

```python
# boş verilere baktık
movies.isnull().sum() 
```

```python
# boş değer içeren satırları silelim kalıcı olarak
movies.dropna(inplace=True)
```

```python
# artık boş değer içeren sütun kalmadı
movies.isnull().sum() 
```

```python
# ilk satırı şöyle döndürürüz: movies.iloc[0]
# ilk satırdaki genres sütununu döndürelim
movies.iloc[0].genres
```

```python
# satırı şöyle getirelim:
# ["Action","Adventure","Fantasy",Science Fiction"]
```

```python
# satırın öyle görünmesi için fonksiyon yazalım
# her seferinde sözlükten bir tanesi alınıp value değerini listeye eklicez
# ast.literal_eval(obj) → obj adlı string’i gerçek Python nesnesine (liste/sözlük gibi) çevirir.
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
movies["genres"]=movies["genres"].apply(convert)
```

```python
movies.head()
```

```python
# şimdi aynı fonksiyonu keywords için de uygulayalım
movies["keywords"]=movies["keywords"].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i["name"])
            counter+=1
        else:
            break
    return L
```

```python
movies["cast"].apply(convert3)
```

```python
movies["cast"]=movies["cast"].apply(convert3)
```

```python
movies.head()
```

```python
# sadece yönetmenleri alalım
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i["job"]=="Director":
            L.append(i["name"])
            break
    return L
```

```python
movies["crew"]=movies["crew"].apply(fetch_director)
```

```python
movies.head()
```

```python
movies["overview"].apply(lambda x:x.split())
```

```python
# overview'ı boşluklara göre ayırarak listeye çevirdik
movies["overview"]=movies["overview"].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# biz dönüşümlerimizi şu dört sütun üzerinde uygulayacağız:
# genres,keywords,cast,crew
```

```python
# şimdi gelecekte hatalara yol açmamak için aradaki boşlukları kaldırmamız gerek:
# "Sam Worthington" ---> "SamWorthington"
movies["genres"]=movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"]=movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"]=movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"]=movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
# böylece tüm veriler aynı formatta oldu
```

```python
# şimdi etiketler diye yeni bir sütun oluşturalım
movies["tags"]=movies["overview"]+movies["genres"]+movies["keywords"]+movies["cast"]+movies["crew"]
```

```python
movies.head()
```

```python
# tags oluştuğu için artık diğer şeylere gerek kalmadı sadece gereklileri alalım
new_df=movies[["movie_id","title","tags"]]
```

```python
new_df
```

```python
# şimdi tags kısmını birleştirelim, boşlukla ayrılmış olsun
new_df["tags"]=new_df["tags"].apply(lambda x:" ".join(x))
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```
